```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# гистограмма
rng = np.random.default_rng(1)
data = rng.normal(size=1000)
plt.hist(data,bins=30,histtype='step', alpha=0.5, edgecolor='red')
plt.show()
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

args = dict(alpha=0.3, bins=40)
plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)
plt.show()
```

```python
print(np.histogram(x1, bins=1))
print(np.histogram(x1, bins=2))
print(np.histogram(x1, bins=40))
```

```python
# двумерные гистограммы
mean = [0,0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
plt.hist2d(x, y, bins=30)
plt.hexbin(x, y)
cb = plt.colorbar()
cb.set_label('Point in interval')
```

```python
print(np.histogram2d(x, y, bins=10))
```

```python
# легенда
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), label='sin')
ax.plot(x, np.cos(x), label='cos')
ax.axis('equal')
ax.legend(frameon=False, fancybox=True, shadow=True)
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
y = np.sin(x[:, np.newaxis]+np.pi * np.arange(0,2, 0.5))
lines = plt.plot(x, y)
plt.legend(lines, ['1', 'second', 'третий', '4ый'], loc='upper left')
```

```python
cities = pd.read_csv('./data/california_cities.csv')
```

```python
lat, lon, pop, area = cities['latd'], cities['longd'], \
cities['population_total'], cities['area_total_km2']
plt.scatter(lon, lat, c=np.log10(pop), s=area)
plt.xlabel('широта')
plt.ylabel('долгота')
plt.colorbar()
plt.clim(3, 7)
plt.scatter([], [], c='k', alpha=0.5, s=100, label='100 km2')
plt.scatter([], [], c='k', alpha=0.5, s=300, label='300 km2')
plt.scatter([], [], c='k', alpha=0.5, s=500, label='500 km2')
plt.legend(labelspacing=10, frameon=False)

plt.legend()
```

```python
fig, ax = plt.subplots()
lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)
for i in range(4):
    lines += ax.plot(x, np.sin(x-i+np.pi/2), styles[i])
ax.axis('equal')
ax.legend(lines[:2], ['l1', 'l2'], loc = 'upper left')
leg = mpl.legend.Legend(ax, lines[1:], ['l2', 'l3', 'l4'], loc = 'lower left')

ax.add_artist(leg)
plt.show()
```

```python
# простые сетки
x = np.linspace(0, 10, 100)  
fig = plt.figure(figsize=(10, 6))

fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.plot(np.sin(x + np.pi/4 * i))
plt.show()
```
