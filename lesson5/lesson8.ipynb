```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z1 = np.linspace(0, 15, 1000) 
y1 = np.cos(z1)
x1 = np.sin(z1)

ax.plot3D(x1, y1, z1, 'green')

z2 = 15*np.random.random(100)  
x2 = np.cos(z2) + 0.1*np.random.random(100)
y2 = np.sin(z2) + 0.1*np.random.random(100)

ax.scatter3D(x2, y2, z2, c=z2, cmap='Greens')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

ax.contour3D(X, Y, Z, 40, cmap='binary')

ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.set_zlabel('z')


ax.view_init(45) 

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

ax.contour3D(X, Y, Z, 40, cmap='binary')

ax.set_xlabel('x') # Оси
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot_wireframe(X, Y, Z)

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

ax.contour3D(X, Y, Z, 40, cmap='binary')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8*np.pi, 40)

R, Theta = np.meshgrid(r, theta)

X = r * np.sin(Theta)
Y = r * np.cos(Theta)
Z = f(X, Y)

ax.plot_surface(X, Y, Z, rstride=1, cmap='viridis', edgecolor='none')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

theta = 2 * np.pi + np.random.random(1000)
r = 6 * np.random.random(1000)

x = r * np.sin(theta)
y = r * np.cos(theta)
z = f(x,y)

ax.scatter(x, y, z, c=z, cmap='viridis')

ax.plot_trisurf(x,y,z,cmap='viridis')
```

```python
import seaborn as sns
# - DataFrame (Matplotlib с Pandas)
# - более высокоуровневый

data = np.random.multivariate_normal([0,0], [[5,2], [2,2]], size = 2000)
data = pd.DataFrame(data, columns=['x','y'])
print(data.head())

fig = plt.figure()
plt.hist(data['x'], alpha = 0.5)
plt.hist(data['y'], alpha = 0.5)

fig = plt.figure()
sns.kdeplot(data=data)

fig = plt.figure()
sns.kdeplot(data=data, fill=True) # shade

plt.show()
```
