```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 20, 5)
y1 = np.array([1, 7, 3, 6, 11])
y2 = np.array([4, 3, 1, 8, 12])

plt.plot(x, y1, 'ro-', label="line 1")  
plt.plot(x, y2, 'gD-.', label="line 2")  

plt.legend()  
plt.show()
```

```python
x = np.linspace(1, 5, 5)
y1 = np.random.randint(7, size=5)
y2 = [(i-3)**2+3 for i in range(1, 6)]  
y3 = [-(i-3)**2-1 for i in range(1, 6)]    

fig = plt.figure(figsize=(6, 6))  
grid = plt.GridSpec(2, 2, figure=fig)  

ax1 = fig.add_subplot(grid[0, :])  
ax1.plot(x, y1)

ax2 = fig.add_subplot(grid[1, 0])
ax2.plot(x, y2)

ax3 = fig.add_subplot(grid[1, 1])
ax3.plot(x, y3)

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 5, 300)  
y = np.cos(3 * x)          
y_with_nan = np.where(y > -0.5, y, np.nan)

plt.plot(x, y_with_nan, color='blue')
plt.ylim(-1, 1)
plt.show()
```
