```python
# сценарий
# командная оболочка IPython
# Jupyter

import matplotlib.pyplot as plt
import numpy as np

# plt.show() - запускается только 1 раз


x = np.linspace(1, 10, 100)
plt.plot(x, np.sin(x))
#plt.show()

plt.plot(x, np.cos(x))
plt.show() # для второго графика второй раз надо написать иначе будет только первый
```

```python
# jupyter
# matplotlib inline - статическая картинка
# matplotlib notebook - интерактивные графики
# fig.savefig('') - сохранение графика в пнг
```

```python
# два способа вывода графиков: - матлаб-подобный стиль, объектно-ориентированный стиль

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
plt.show()
# этот способ содержит объекты (ск, тексты, метки), ax:Axes - ск, прямоугольник, деления, метки
```

```python
# задаем цвет: константами (blue), rgb, html, градация серого от 0 до 1 - 0.14, 0.1 и тд
# можно так
fig = plt.figure()
ax = plt.axes() #так будет 1 график
ax.plot(x, np.sin(x), color='blue')
ax.plot(x, np.sin(x-1))
ax.plot(x, np.sin(x-2))
ax.plot(x, np.sin(x-3))
plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.xlabel('x')
plt.ylabel('sinx')
plt.title('sin')
plt.plot(x, np.sin(x), '-g')

plt.subplot(3, 1, 2)
plt.xlabel('x')
plt.ylabel('cosx')
plt.title('cos')
plt.plot(x, np.cos(x), '-b')
#plt.legend()

plt.subplot(3, 1, 3)
plt.xlabel('x')
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.cos(x), '-b')
plt.axis('equal')
plt.subplots_adjust(hspace=0.5)
```

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), '>', color='green')
```

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), '--p', markersize=15, linewidth=4, markerfacecolor='white', markeredgecolor='gray', markeredgewidth=2)
```

```python
rng = np.random.default_rng(10)
colors = rng.random(30)
sizes = 100 * rng.random(30)
plt.scatter(x, np.sin(x), marker='o', c=colors, s=sizes)
# если больше 1000 точек то лучше plot
```

```python
# визуализация погрешности
x = np.linspace(0, 10, 30)
dy = 0.4
y = np.sin(x) + dy * np.random.randn(30)
plt.errorbar(x, y, yerr=dy)
plt.fill_between(x, y-dy, y+dy, color='red')
```

```python
def f(x, y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 30)
y = np.linspace(0, 5, 30)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)
c = plt.contour(X, Y, Z)
plt.clabel(c)
#plt.colorbar()
plt.imshow(Z, extent=[0, 5, 0, 5], cmap='RdGy',
           interpolation='gaussian', origin='upper', aspect='equal')
plt.show()
```
